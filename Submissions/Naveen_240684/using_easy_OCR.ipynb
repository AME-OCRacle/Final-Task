```python
import jiwer
text= " ".join(text)
reference = "this is a handwritten example write as good as you can."

wer = jiwer.wer(reference, text)

cer = jiwer.cer(reference, text)

print(f"WER: {wer:.2f}")
print(f"CER: {cer:.2f}")
```
